# 3.d- Diviseur de tension et de courant
Notre circuit est composé d'une résistance de 1 $\Omega$ en série avec deux résistances de 2 $\Omega$ et $1\Omega$, le tout alimenté par une source de tension de 9V.
Voici les étapes de notre démarche pour déterminer le potentiel et le courant partout dans le circuit.


Commençons par trouver la résistance équivalente des 3 résistances.
$$R_{eq} = 1\Omega +\frac{1}{\frac{1}{1\Omega}+\frac{1}{2\Omega}} = \frac{5}{3}\Omega$$
le courant passant par la source et la première résistance est donc donné par:
$$ I=\frac{V}{R}=\frac{9V}{\frac{5}{3}\Omega}= 5.4 A$$
La différence de potentiel autour de la première résistance est donc:
$$ V=RI=1\Omega5.4A = 5.4V$$
En posant le potentiel de la borne positive à 4.5 V et celui de la borne négative à -4.5V, on trouve que le potentiel électrique du noeud supérieur est de -0.9 V. Le reste du circuit à donc une différence de potentiel de 3.6V. On peut donc trouver le courant dans chaque branche restante. Pour la résistance de 2$\Omega$:

$$ I = \frac{V}{R} = \frac{3.6V}{2\Omega}=1.8A$$
Pour celle de $1\Omega$:
$$ I = \frac{V}{R} = \frac{3.6V}{1\Omega}=3.6A$$



 

```python
world = World(shape=(91, 91))
wires = [
    Wire(start=(20, 71), stop=(30, 71), current=Current(x=5.4, y=0), voltage=4.5),
    Wire(start=(30, 71), stop=(46, 71), current=Current(x=5.4, y=0), voltage=-0.9),
    Wire(start=(46, 71), stop=(71, 71), current=Current(x=1.8, y=0), voltage=-0.9),
    Wire(start=(71, 71), stop=(71, 46), current=Current(x=0, y=-1.8), voltage=-0.9),
    Wire(start=(71, 46), stop=(71, 20), current=Current(x=0, y=-1.8), voltage=-4.5),
    Wire(start=(71, 20), stop=(46, 20), current=Current(x=-1.8, y=0), voltage=-4.5),
    Wire(start=(46, 20), stop=(30, 20), current=Current(x=-5.4, y=0), voltage=-4.5),
    Wire(start=(30, 20), stop=(20, 20), current=Current(x=-5.4, y=0), voltage=4.5),
    Wire(start=(20, 20), stop=(20, 71), current=Current(x=0, y=5.4), voltage=4.5),
    Wire(start=(46, 71), stop=(46, 46), current=Current(x=0, y=-3.6), voltage=-0.9),
    Wire(start=(46, 46), stop=(46, 20), current=Current(x=0, y=-3.6), voltage=-4.5)
]
circuit = Circuit(wires=wires)
world.place(circuit)
world.compute(50000)


num = 'd'
world.show_wires_voltage(Titre=f'Potentiel électrique initial du circuit 3.{num} (en V)')
world._potential.show(title=f'Potentiel électrique autour du circuit 3.{num} (en V)', x_label='x [m]', y_label = 'y [m]')
world.show_electric_field(Titre=f'Champs électrique autour du circuit 3.{num} (en V/m)')
world.show_magnetic_field(Titre=f'Champs magnétique autour du circuit 3.{num} (en T $\hat z$)')
world.show_energy_flux(Titre = f'Vecteurs de Poynting autour du circuit 3.{num} (en W/m$^2$)')
```
